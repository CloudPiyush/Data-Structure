```python
class TreeNode:

    def __init__(self,Data) -> None:
        self.Data = Data
        self.Children = []
        self.Parent = None

    def Get_Length(self):
        Level = 0
        P = self.Parent
        while P:
            Level += 1
            P = P.Parent
        return Level
    
    def Get_Skills(self,skill):
        skill.Parent = self
        self.Children.append(skill)

    def Show_Skills(self):
        Space = ' ' * self.Get_Length() *3
        Prefix = Space + '|__' if self.Parent else ''
        print(Prefix + self.Data)
        if self.Children:
            for skill in self.Children:
                skill.Show_Skills()

def Get_Details():

    Root = TreeNode('Piyush skills')

    Inheritence = TreeNode('Inheritence')
    Inheritence.Get_Skills(TreeNode('single Inheritence'))
    Inheritence.Get_Skills(TreeNode('Multiple Inheritence'))
    Inheritence.Get_Skills(TreeNode('Multilevel Inheritence'))
    Inheritence.Get_Skills(TreeNode('hybrid Inheritence'))

    Operator_OverLoading = TreeNode('Operator Overloading')
    Operator_OverLoading.Get_Skills(TreeNode('Early Binding'))

    Operator_Overriding = TreeNode('Operator Overriding')
    Operator_Overriding.Get_Skills(TreeNode('Late Binding'))

    Polymorphism = TreeNode('Polymorphism')
    Polymorphism.Get_Skills(Operator_OverLoading)
    Polymorphism.Get_Skills(Operator_Overriding)

    object_oriented_programming = TreeNode('Object Oriented Programming')
    object_oriented_programming.Get_Skills(TreeNode('class'))
    object_oriented_programming.Get_Skills(TreeNode('Object'))
    object_oriented_programming.Get_Skills(TreeNode('Encapsulation'))
    object_oriented_programming.Get_Skills(TreeNode('Constructor'))
    object_oriented_programming.Get_Skills(Inheritence)
    object_oriented_programming.Get_Skills(Polymorphism)

    Python = TreeNode('Python Programming')
    Python.Get_Skills(TreeNode('Numpy'))
    Python.Get_Skills(TreeNode('Pandas'))
    Python.Get_Skills((object_oriented_programming))

    ML = TreeNode('Machine Learning')
    ML.Get_Skills(TreeNode('Supervised Learning'))
    ML.Get_Skills(TreeNode('Unsupervised Learning'))
    ML.Get_Skills(TreeNode('Renforcement Learning'))

    AI = TreeNode('Artificial Intelligence')
    AI.Get_Skills(ML)
    AI.Get_Skills(TreeNode('Deep Learning'))
    AI.Get_Skills(TreeNode('Natural Language Processs'))
    AI.Get_Skills(TreeNode('Neural Networks'))


    Root.Get_Skills(Python)
    Root.Get_Skills(AI)
    Root.Show_Skills()
```

```python
if __name__ == "__main__":
    Get_Details()
```
